### Model Training - Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Assigning the test size as 30% of the data set
df_train , df_test = train_test_split(df, test_size= 0.30)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
X_train = df_train.drop(columns = ["Target", "Species"]).values
```

```python
y_train = df_train["Target"].values
```

```python
# Fitting the model
model = LogisticRegression(max_iter=200)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train) * 100
```

Using the same training model from the train test split section to predect accuracy is faulty. The model is basically overfitted, a better practice is to use a validation set on the model that was created earlier.

```python
#Making a validation set, where v is for validation

xt, xv, yt, yv = train_test_split(X_train, y_train, test_size=0.30)
```

```python
xt.shape
```

```python
model.fit(xt, yt)
```

```python
model.score(xv, yv)*100
```

Using a validation set, increased the prediction accuracy to 100%

### Cross - Validation

K-fold validation tests different portions of the entire dataset to have a more thorough accuracy.

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
acc = cross_val_score(model, xv, yv, cv=5)
```

```python
acc
```

Seeing as the prediction is as low as 83% accuracy, we can double check to see what values are being incorrectly classified.

```python
np.mean(acc)
```

```python
y_pred = cross_val_predict(model, X_train, y_train, cv=5)
```

```python
y_pred == y_train
```

This array can be put into a mask, in order to pinpoint which dataset in particular, was incorrectly classified.

```python
correct_pred_mask = y_pred == y_train
```

```python
incorrect_pred_mask = ~correct_pred
```

```python
incorrect_pred_mask #this just inverts the boolean value to make the missed prediction "true"
```

```python
X_train[incorrect_pred_mask]
```

Let's find that data point from the original dataframe

```python
df_predicted = df_train.copy()
```

```python
df_predicted["Is it Correct?"]= correct_pred_mask
```

```python
df_predicted.head()
```

```python
df_predicted["Predict"]= y_pred
```

```python
df_predicted["Predicted Labels"] = df_predicted["Predict"].map({0:'setosa' , 1:'versicolor', 2:'virginica'})
```

```python
df_predicted.head()
```

```python
np.where(df_predicted["Is it Correct?"] == False)
```

```python
sns.scatterplot(x="petal length (cm)", y= "petal width (cm)", hue="Predicted Labels", data= df_predicted)
```

```python
sns.scatterplot(x="petal length (cm)", y= "petal width (cm)", hue="Species", data= df_predicted)
```

Comparing these two graphs, it looks like just one versicolor iris (blue dot) was predicted to be orange in the model that was tested. Not bad for a prediction.
